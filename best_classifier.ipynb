### Load Data From CSV File  


```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
from sklearn import preprocessing
%matplotlib inline
```

```python
dataset = pd.read_csv('loan_train.csv')
dataset.head(5)
```

```python
dataset['due_date'] = pd.to_datetime(dataset['due_date'])
dataset['effective_date'] = pd.to_datetime(dataset['effective_date'])
dataset.head(10)
```

```python
dataset['loan_status'].value_counts()
```

```python
dataset['dayofweek'] = dataset['effective_date'].dt.dayofweek
```

```python
dataset['weekend'] = dataset['dayofweek'].apply(lambda x: 1 if (x>3)  else 0)
dataset.head()
```

```python
# convert categorical data to numerical
dataset.groupby(['Gender'])['loan_status'].value_counts(normalize=True)
```

```python
#  let convert male to 0 and female to 1

dataset['Gender'].replace(to_replace=['male','female'], value=[0,1],inplace=True)
dataset.head()
```

#  one hot encoding

```python
dataset.groupby(['education'])['loan_status'].value_counts(normalize=True)
```

## Use one hot encoding technique to conver categorical varables to binary variables and append them to the feature Data Frame 

```python
Feature = dataset[['Principal','terms','age','Gender','weekend']]
Feature = pd.concat([Feature,pd.get_dummies(dataset['education'])], axis=1)
Feature.drop(['Master or Above'], axis = 1,inplace=True)
Feature.head()
```

```python
#  Feature selection
x=Feature
x[0:5]
```

```python
# Replacing "PAIDOFF" and "COLLECTION" strings with 0 and 1 integers because some models can only work with integers or floats
```

```python
dataset['loan_status'].replace(to_replace=['PAIDOFF','COLLECTION'], value=[0,1],inplace=True)
dataset.head()
```

```python
y = dataset['loan_status'].values
y[0:20]
```

## Doing Train test split -

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.40, random_state=5)
```

```python
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

## Normalize Data 

```python
# (technically should be done after train test split )
X_train = preprocessing.StandardScaler().fit(X_train).transform(X_train)
X_test = preprocessing.StandardScaler().fit(X_test).transform(X_test)
X_train[0:5]
```

# Classification 

# K Nearest Neighbor(KNN)
You should find the best k to build the model with the best accuracy.  


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
```

```python
k = 3
#We fit the model:
kNN_model = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
kNN_model
y_pred = kNN_model.predict( X_test )

#Find the max value
accuracies = {}
k_max = 1
acc_max = 0
for k in range(1, 10):
    kNN_model = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    y_pred = kNN_model.predict( X_test )
    accuracies[k] = accuracy_score(y_test, y_pred)
    print(k, accuracies[k])
```

```python
print("We take k = 7")
knn_model = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train)

print("Train set Accuracy (Jaccard): ", accuracy_score(y_train, knn_model.predict(X_train)))
print("Test set Accuracy (Jaccard): ", accuracy_score(y_test, knn_model.predict(X_test)))

print("Train set Accuracy (F1): ", f1_score(y_train, knn_model.predict(X_train), average='weighted'))
print("Test set Accuracy (F1): ", f1_score(y_test, knn_model.predict(X_test), average='weighted'))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
for d in range(1,10):
    dt = DecisionTreeClassifier(criterion = 'entropy', max_depth = d).fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    print(d, accuracy_score(y_test, y_pred))
```

```python
print("We take depth = 2")
dt = DecisionTreeClassifier(criterion="entropy", max_depth=2).fit(X_train, y_train)

print("Train set Accuracy (Jaccard): ", accuracy_score(y_train, dt.predict(X_train)))
print("Test set Accuracy (Jaccard): ", accuracy_score(y_test, dt.predict(X_test)))

print("Train set Accuracy (F1): ", f1_score(y_train, dt.predict(X_train), average='weighted'))
print("Test set Accuracy (F1): ", f1_score(y_test, dt.predict(X_test), average='weighted'))
```

# Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
for k in ('linear', 'poly', 'rbf','sigmoid'):
    svm_model = SVC( kernel = k).fit(X_train,y_train)
    svm_yhat = svm_model.predict(X_test)
    print("For kernel: {}, the f1 score is: {}".format(k,f1_score(y_test,svm_yhat, average='weighted')))
```

```python
print("So we choose rbf")
svm_model = SVC( kernel = 'rbf').fit(X_train,y_train)


print("Train set Accuracy (Jaccard): ", accuracy_score(y_train, svm_model.predict(X_train)))
print("Test set Accuracy (Jaccard): ", accuracy_score(y_test, svm_model.predict(X_test)))

print("Train set Accuracy (F1): ", f1_score(y_train, svm_model.predict(X_train), average='weighted'))
print("Test set Accuracy (F1): ", f1_score(y_test, svm_model.predict(X_test), average='weighted'))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
for k in ('lbfgs', 'saga', 'liblinear', 'newton-cg', 'sag'):
    lr_model = LogisticRegression(C = 0.01, solver = k).fit(X_train, y_train)
    lr_yhat = lr_model.predict(X_test)
    y_prob = lr_model.predict_proba(X_test)
    print('When Solver is {}, logloss is : {}'.format(k, log_loss(y_test, y_prob)))
```

```python
k = 'liblinear'
lr_model = LogisticRegression(C = 0.01, solver = k).fit(X_train, y_train)

print("Train set Accuracy (Jaccard): ", accuracy_score(y_train, lr_model.predict(X_train)))
print("Test set Accuracy (Jaccard): ", accuracy_score(y_test, lr_model.predict(X_test)))

print("Train set Accuracy (F1): ", f1_score(y_train, lr_model.predict(X_train), average='weighted'))
print("Test set Accuracy (F1): ", f1_score(y_test, lr_model.predict(X_test), average='weighted'))
```
